# ReDub usage examples

This notebook shows the core authoring flow: build JSX, compile it to an AST, and serialize it to TTML2 / DAPT XML.

```javascript
import { Redub, compile, serialize } from "redub";

const document = compile(
  <Redub xmlLang="en">
    <div begin="10s" end="13s">
      <p>
        <span begin="0s">Hello</span>
        <span begin="1.5s"> world</span>
      </p>
    </div>
  </Redub>
);

document;
```

```javascript
serialize(document);
```

## Metadata example

```javascript
import { Agent, Pronunciation } from "redub";

const metadataDocument = compile(
  <Redub
    xmlLang="en"
    langSrc="fr"
    scriptType="preRecording"
    scriptRepresents={["audio.dialogue"]}
  >
    <Redub.Head>
      <Redub.Metadata>
        <Agent id="character_1" type="character" alias="ASSANE" />
        <Pronunciation target="SQL" alias="sequel" />
      </Redub.Metadata>
    </Redub.Head>
    <div {...({ agent: "character_1", represents: "audio.dialogue" } as object)}>
      <p>
        <span>Hello world</span>
      </p>
    </div>
  </Redub>
);

serialize(metadataDocument);
```

## Frame timing example

```javascript
const frameTimedDocument = compile(
  <Redub>
    <div {...({ from: 300, to: 390, fps: 30 } as object)}>
      <p>
        <span>Frame-based timing</span>
      </p>
    </div>
  </Redub>
);

serialize(frameTimedDocument);
```
