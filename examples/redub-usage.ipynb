# ReDub usage examples

This notebook shows the core authoring flow: build a React element tree, compile it to an AST, and serialize it to TTML2 / DAPT XML.

The examples use `React.createElement(...)` so they can run in a plain JavaScript kernel without JSX transpilation.

```javascript
import React from "react";
import { Redub, compile, serialize } from "redub";

const e = React.createElement;

const document = compile(
  e(
    Redub,
    { xmlLang: "en" },
    e(
      "div",
      { begin: "10s", end: "13s" },
      e(
        "p",
        null,
        e("span", { begin: "0s" }, "Hello"),
        e("span", { begin: "1.5s" }, " world")
      )
    )
  )
);

document;
```

```javascript
serialize(document);
```

## Metadata example

```javascript
import { Agent, Pronunciation } from "redub";

const metadataDocument = compile(
  e(
    Redub,
    {
      xmlLang: "en",
      langSrc: "fr",
      scriptType: "preRecording",
      scriptRepresents: ["audio.dialogue"],
    },
    e(
      Redub.Head,
      null,
      e(
        Redub.Metadata,
        null,
        e(Agent, { id: "character_1", type: "character", alias: "ASSANE" }),
        e(Pronunciation, { target: "SQL", alias: "sequel" })
      )
    ),
    e(
      "div",
      { agent: "character_1", represents: "audio.dialogue" },
      e("p", null, e("span", null, "Hello world"))
    )
  )
);

serialize(metadataDocument);
```

## Frame timing example

```javascript
const frameTimedDocument = compile(
  e(
    Redub,
    null,
    e(
      "div",
      { from: 300, to: 390, fps: 30 },
      e("p", null, e("span", null, "Frame-based timing"))
    )
  )
);

serialize(frameTimedDocument);
```
